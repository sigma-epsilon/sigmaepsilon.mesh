# Voxelizing a Tetrahedral Mesh

```python
from sigmaepsilon.mesh.downloads import download_stand

mesh = download_stand(read=True)
mesh.plot(notebook=True)
print("Volume: ", mesh.volume())
```

```python
from sigmaepsilon.mesh.voxelize import voxelize_TET4_H8

coords_TET4 = mesh.coords()
topo_TET4 = mesh.topology().to_numpy()
shape = (50, 50, 50)
hexa_mesh = voxelize_TET4_H8(coords_TET4, topo_TET4, shape=shape, k_max=10)
```

```python
from sigmaepsilon.mesh.cells import H8 as CellData
from sigmaepsilon.mesh import PolyData, PointData

mesh_H8 = PolyData(
    pd=PointData(coords=hexa_mesh[0]),
    cd=CellData(topo=hexa_mesh[1]),
)
mesh_H8.plot(notebook=True)
print("Volume: ", mesh_H8.volume())
```
