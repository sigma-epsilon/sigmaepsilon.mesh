# Voxelizing a Triangular Mesh

```python
from sigmaepsilon.mesh.downloads import download_bunny_coarse

mesh = download_bunny_coarse(read=True)
mesh.plot(notebook=True)
```

```python
from sigmaepsilon.mesh.voxelize import voxelize_T3_H8

coords_T3 = mesh.coords()
topo_T3 = mesh.topology().to_numpy()
shape = (50, 50, 50)
hexa_mesh = voxelize_T3_H8(coords_T3, topo_T3, shape=shape, k_max=10)
```

```python
from sigmaepsilon.mesh.cells import H8 as CellData
from sigmaepsilon.mesh import PolyData, PointData

mesh_H8 = PolyData(
    pd=PointData(coords=hexa_mesh[0]),
    cd=CellData(topo=hexa_mesh[1]),
)
mesh_H8.plot(notebook=True)
```
