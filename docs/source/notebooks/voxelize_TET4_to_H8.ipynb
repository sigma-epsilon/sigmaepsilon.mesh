# Voxelizing a Tetrahedral Mesh

```python
import numpy as np

coords_TET4 = np.array([[0, 0, 0], [1, 0, 0], [0, 1, 0], [0, 0, 1]]).astype(float)
topo_TET4 = np.array([[0, 1, 2, 3]])
tetra_mesh = (coords_TET4, topo_TET4)
```

```python
from sigmaepsilon.mesh import PolyData, PointData
from sigmaepsilon.mesh.space import StandardFrame
from sigmaepsilon.mesh.cells import TET4 as CellData

frame = StandardFrame(dim=3)
mesh = PolyData(
    pd=PointData(coords=tetra_mesh[0], frame=frame),
    cd=CellData(topo=tetra_mesh[1], frames=frame),
)
mesh.plot(notebook=True)
```

```python
from sigmaepsilon.mesh.voxelize import voxelize_TET4_H8

shape = (10, 10, 10)
hexa_mesh = voxelize_TET4_H8(coords_TET4, topo_TET4, shape=shape, k_max=1)
```

```python
from sigmaepsilon.mesh.cells import H8 as CellData

mesh = PolyData(
    pd=PointData(coords=hexa_mesh[0], frame=frame),
    cd=CellData(topo=hexa_mesh[1], frames=frame),
)
mesh.plot(notebook=True)
```
